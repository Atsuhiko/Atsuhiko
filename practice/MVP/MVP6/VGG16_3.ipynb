### 予測の関数化

```python
def Predict_Dogs_Cats():
    filenames = os.listdir("./images")
    sample = random.choice(filenames)
    img = load_img("./images/"+sample, target_size=IMAGE_SIZE)
    img_org = img
    img = np.asarray(img)
    img = np.expand_dims(img, axis=0)
    predict = model.predict(img)
    plt.imshow(img_org)

    if predict[0][0] > predict[0][1]:
            print("この画像はネコです")
    else:
            print('この画像は犬です')
    
    print("dog: " + predict[0][1].astype(str))
    print("cat: " + predict[0][0].astype(str))
```

```python
Predict_Dogs_Cats()
```
